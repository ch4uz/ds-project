This notebook must be run before the other preperation tasks, as it removes redundant (duplicate) variables with different representation, and leaky variables that shouldn't be known at the time of prediction. The removals are based on data profiling/inspection, domain knowledge and common sense.

```python
import pandas as pd
from pathlib import Path

input_path = Path("../../data/raw/Combined_Flights_2022.csv")

df = pd.read_csv(input_path)
```

```python
leaky_cols_to_drop = [
    "Diverted",
    "DivAirportLandings"
    "DepTime",
    "DepDelayMinutes",
    "DepDelay",
    "ArrTime",
    "ArrDelayMinutes",
    "AirTime",
    "ActualElapsedTime",
    "DepDel15",
    "DepartureDelayGroups",
    "TaxiOut",
    "TaxiIn",
    "WheelsOff",
    "WheelsOn",
    "ArrDelay",
    "ArrDel15",
    "ArrivalDelayGroups",   
]

df.drop(columns=leaky_cols_to_drop, errors="ignore", inplace=True)
```

```python
redundant_cols_to_drop = [
     "DOT_ID_Marketing_Airline", # same as Marketing_Airline_Network
     "IATA_Code_Marketing_Airline", # same as Marketing_Airline_Network
     "DOT_ID_Operating_Airline", # same as Operating_Airline
     "IATA_Code_Operating_Airline", # same as Operating_Airline
     "OriginAirportID", # same as Origin
     "OriginAirportSeqID", # same as Origin
     "OriginStateFips", # same as OriginState
     "OriginStateName", # same as OriginState
     "OriginWac", # same as OriginState
     "DestAirportID", # same as Dest
     "DestAirportSeqID", # same as Dest
     "DestStateFips", # same as DestState
     "DestStateName", # same as DestState
     "DestWac", # same as DestState
]

df.drop(columns=redundant_cols_to_drop, errors="ignore", inplace=True)
```

```python
output_path = input_path.with_name(input_path.stem + "_cleaned.csv")
df.to_csv(output_path, index=False)
```
