This file samples from the flight data and creates separate train and test data files with rows in chronological order. 

Change file names to reuse.

```python
import pandas as pd
```

```python
# Load dataset from file
data = pd.read_csv("../data/prepared/mvi/flight_mvi.csv")
data.head()
print(data.shape)
```

```python
# Sample 100 000 rows for quicker experimentation
data = data.sample(n=100_000, random_state=42)
data.shape
```

```python
# Sort the dataset by 'FlightDate' to maintain temporal order
data = data.sort_values(by="FlightDate").reset_index(drop=True)
```

```python
# Do a train-test split based on time
train_size = int(0.7 * len(data))
train_data = data.iloc[:train_size]
test_data = data.iloc[train_size:]

print(f"Train data shape: {train_data.shape}")
print(f"Test data shape: {test_data.shape}")
```

```python
# Save train and test datasets
train_data.to_csv("../data/prepared/mvi/flight_mvi_sampled_train.csv", index=False)
test_data.to_csv("../data/prepared/mvi/flight_mvi_sampled_test.csv", index=False)
```
