This file samples from the <b>encoded</b> flight data and creates separate train and test data files.

```python
import pandas as pd
```

```python
# Load dataset from file
data_enc = pd.read_csv("../data/prepared/flight_enc.csv")
data_enc.head()
print(data_enc.shape)
```

```python
# Sample 500 000 rows for quicker experimentation
data_enc = data_enc.sample(n=500_000, random_state=42)
data_enc.shape
```

```python
# Sort the dataset by 'FlightDate' to maintain temporal order
data_enc = data_enc.sort_values(by="FlightDate").reset_index(drop=True)
```

```python
# Do a train-test split based on time
train_size = int(0.7 * len(data_enc))
train_data = data_enc.iloc[:train_size]
test_data = data_enc.iloc[train_size:]

print(f"Train data shape: {train_data.shape}")
print(f"Test data shape: {test_data.shape}")
```

```python
# Save train and test datasets
train_data.to_csv("../data/prepared/flight_enc_sampled_train.csv", index=False)
test_data.to_csv("../data/prepared/flight_enc_sampled_test.csv", index=False)
```
